
# Step 2 — Black‑Box Model (XGBoost) for Default Prediction

This notebook trains an XGBoost classifier to forecast **default** (binary `target`) for the course project **Interpretability, Stability, and Algorithmic Fairness (Fall 2025)**.

**What this notebook does**
- Loads the project dataset
- Cleans & preprocesses features (numeric + categorical)
- Trains an XGBoost model with early stopping
- Evaluates with ROC‑AUC, PR‑AUC, F1, confusion matrix
- Saves predictions and the fitted model for later steps

> Note: Columns that could leak information from Step 1 (e.g., `'Predictions'`, `'Predicted probabilities'`) are **dropped** on load.


## 0. Setup

```python

# If needed, install dependencies (uncomment as necessary)
# %pip install -q xgboost scikit-learn pandas numpy matplotlib joblib

import os
import gc
import math
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import (
    roc_auc_score, average_precision_score, precision_recall_curve,
    roc_curve, classification_report, confusion_matrix
)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from xgboost import XGBClassifier

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Configuration

```python

# --- Paths ---
# Prefer a local file first; fall back to /mnt/data if running in a hosted environment.

DATA_PATH = '/Users/chenzhipeng/Desktop/X/Cours/Interpretability/group_project/Algo_fairness-Group-Project/data/raw/dataproject2025.csv'

# --- Basic run params ---
TEST_SIZE = 0.20            # 80/20 stratified split
USE_SAMPLE = False          # set True for quick tests (e.g., 50_000 rows)
SAMPLE_N = 50_000

# Whether to EXCLUDE sensitive attributes from features (you can toggle later for Step 9 fairness analysis)
EXCLUDE_SENSITIVE = False
SENSITIVE_COLS = [
    "Pct_afro_american"    # protected attribute proxy provided in the dataset
]

print(f"Using DATA_PATH = {DATA_PATH}")
```

## 2. Load data

```python

# Read CSV (large-friendly). We'll downcast floats to float32 to reduce memory.
read_kwargs = dict(low_memory=False)
df = pd.read_csv(DATA_PATH, **read_kwargs)

if USE_SAMPLE and len(df) > SAMPLE_N:
    df = df.sample(SAMPLE_N, random_state=RANDOM_STATE).reset_index(drop=True)

print(df.shape)
display(df.head(3))
print("\nColumns:", list(df.columns))
```

## 3. Basic cleaning & target

```python

# Drop known leak / index columns if present
LEAKY_COLS = [
    "Predictions", "Predicted probabilities", "DP", "dp", 
    "Unnamed: 0", "id", "ID", "index"
]
present_leaky = [c for c in LEAKY_COLS if c in df.columns]
df = df.drop(columns=present_leaky, errors="ignore")

# Target
assert "target" in df.columns, "Expected 'target' column not found."
df["target"] = df["target"].astype(int)

# Optional: exclude sensitive columns for Step 2 (toggle above)
if EXCLUDE_SENSITIVE:
    drop_sens = [c for c in SENSITIVE_COLS if c in df.columns]
    if drop_sens:
        df = df.drop(columns=drop_sens)

print("After drops:", df.shape)
df.head(2)
```

## 4. Feature engineering

```python

def parse_emp_length(val):
    '''
    Convert employment length strings to numeric years.
    Examples: '10+ years'->10, '< 1 year'->0.5, '3 years'->3, NaN->np.nan
    '''
    if pd.isna(val):
        return np.nan
    s = str(val).strip().lower()
    if s in {"< 1 year", "less than 1 year", "<1 year"}:
        return 0.5
    if s in {"10+ years", "10+ yrs", "10+yr"}:
        return 10.0
    # Extract any leading integer
    for tok in s.replace("+","").split():
        try:
            return float(int(tok))
        except:
            continue
    try:
        return float(s)
    except:
        return np.nan

def parse_loan_duration(val):
    '''
    Convert '36 months' -> 36, '60 months' -> 60, already numeric -> itself.
    '''
    if pd.isna(val):
        return np.nan
    if isinstance(val, (int, float)):
        return float(val)
    s = str(val).strip().lower().replace("months","").replace("month","").strip()
    try:
        return float(s)
    except:
        return np.nan

# Apply parsers if columns exist
if "emp_length" in df.columns:
    df["emp_length_parsed"] = df["emp_length"].apply(parse_emp_length)
if "loan duration" in df.columns:
    df["loan_duration_parsed"] = df["loan duration"].apply(parse_loan_duration)

# If 'issue_d' is a date-like string or year, keep numeric
if "issue_d" in df.columns:
    # If looks numeric (e.g., 2013.0), keep as float; else parse to ordinal
    if df["issue_d"].dtype == object:
        try:
            d = pd.to_datetime(df["issue_d"], errors="coerce")
            df["issue_d_ordinal"] = d.map(lambda x: x.toordinal() if pd.notna(x) else np.nan)
        except:
            pass
    else:
        # leave numeric year/month values as-is
        pass

# Identify feature columns
target_col = "target"
all_features = [c for c in df.columns if c != target_col]

# Separate numeric vs categorical by dtype (object -> categorical). We'll skip free-text high-cardinality fields.
high_cardinality_text = {"emp_title"}  # can revisit with hashing later
features = [c for c in all_features if c not in high_cardinality_text]

numeric_cols = [c for c in features if pd.api.types.is_numeric_dtype(df[c])]
categorical_cols = [c for c in features if c not in numeric_cols]

print("Numeric:", len(numeric_cols), "Categorical:", len(categorical_cols))
```

## 5. Train / validation split

```python

X = df[numeric_cols + categorical_cols].copy()
y = df[target_col].copy()

# Ensure numeric types are float32 to reduce memory footprint
for c in numeric_cols:
    X[c] = pd.to_numeric(X[c], errors="coerce").astype("float32")

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, stratify=y, random_state=RANDOM_STATE
)

print("Train:", X_train.shape, " Test:", X_test.shape)
print("Default rate (train):", y_train.mean().round(4), " (test):", y_test.mean().round(4))
```

## 6. Model: XGBoost

```python
# Preprocess
numeric_transformer = SimpleImputer(strategy="median")

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=True))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_cols),
        ("cat", categorical_transformer, categorical_cols),
    ],
    sparse_threshold=0.3  # keep sparse matrices when many categories
)

# Class imbalance handling via scale_pos_weight = (neg/pos)
pos = y_train.sum()
neg = len(y_train) - pos
scale_pos_weight = float(neg / max(pos, 1))

xgb = XGBClassifier(
    n_estimators=2000,
    learning_rate=0.03,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.0,
    reg_lambda=1.0,
    min_child_weight=1.0,
    objective="binary:logistic",
    tree_method="hist",
    random_state=RANDOM_STATE,
    scale_pos_weight=scale_pos_weight,
    n_jobs=-1,
    early_stopping_rounds=50
)

# Early stopping with a validation split
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train, test_size=0.15, stratify=y_train, random_state=RANDOM_STATE
)

model = Pipeline(steps=[
    ("prep", preprocess),
    ("xgb", xgb)
])

# Fit with early stopping (pass eval_set to underlying xgb via fit_params)
# We need to transform X_val with the preprocessor to feed to xgb for eval_set
prep = preprocess.fit(X_tr, y_tr)
X_tr_t = prep.transform(X_tr)
X_val_t = prep.transform(X_val)

xgb.fit(
    X_tr_t, y_tr, # Using your transformed data
    eval_set=[(X_val_t, y_val)],
    verbose=False, # Control verbosity here
  #  callbacks=[EarlyStopping(rounds=50)] # Callback only needs the number of rounds
)

# Wrap back into a single pipeline for convenience (prep is already fitted; xgb is fitted).
from sklearn.base import BaseEstimator, TransformerMixin
class PretrainedPreprocessor(BaseEstimator, TransformerMixin):
    def __init__(self, fitted_preprocessor):
        self.fitted = fitted_preprocessor
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return self.fitted.transform(X)

final_model = Pipeline(steps=[
    ("prep", PretrainedPreprocessor(prep)),
    ("xgb", xgb)
])

try:
    best_ntree = xgb.get_booster().best_ntree_limit
except Exception:
    best_ntree = "n/a"
print("Best n_estimators (limit):", best_ntree)
```

## 7. Evaluation

```python

# Predict
X_test_t = prep.transform(X_test)
proba = xgb.predict_proba(X_test_t)[:, 1]
pred = (proba >= 0.5).astype(int)

# Metrics
roc_auc = roc_auc_score(y_test, proba)
pr_auc = average_precision_score(y_test, proba)
print(f"ROC-AUC: {roc_auc:.4f} | PR-AUC: {pr_auc:.4f}\n")

print(classification_report(y_test, pred, digits=4))

# Confusion matrix
cm = confusion_matrix(y_test, pred)
print("Confusion Matrix:\n", cm)

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, proba)
plt.figure()
plt.plot(fpr, tpr, label=f"ROC-AUC={roc_auc:.3f}")
plt.plot([0,1],[0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve — XGBoost (Step 2)")
plt.legend(loc="lower right")
plt.show()

# PR Curve
prec, rec, _ = precision_recall_curve(y_test, proba)
plt.figure()
plt.plot(rec, prec, label=f"PR-AUC={pr_auc:.3f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve — XGBoost (Step 2)")
plt.legend(loc="lower left")
plt.show()
```

## 8. Feature importance (gain)

```python

# Map feature names through the fitted preprocessor (OHE expands categories)
ohe = prep.named_transformers_["cat"].named_steps["ohe"] if len(categorical_cols) else None
num_feature_names = numeric_cols
cat_feature_names = list(ohe.get_feature_names_out(categorical_cols)) if ohe is not None else []
feature_names = list(num_feature_names) + cat_feature_names

booster = xgb.get_booster()
# XGBoost importance is keyed by f0, f1, ...
score_map = booster.get_score(importance_type="gain")

# Convert to a sorted DataFrame
imp = pd.DataFrame(
    [(int(k[1:]), v) for k,v in score_map.items()],
    columns=["index","gain"]
).sort_values("gain", ascending=False)

imp["feature"] = imp["index"].map(lambda i: feature_names[i] if i < len(feature_names) else f"f{i}")
topk = imp.head(30)[["feature","gain"]]
display(topk)

plt.figure()
plt.barh(topk["feature"][::-1], topk["gain"][::-1])
plt.xlabel("Gain importance")
plt.title("Top features — XGBoost (gain)")
plt.tight_layout()
plt.show()
```

## 9. Save artifacts

```python
import datetime
out_dir = "outputs_step2"
os.makedirs(out_dir, exist_ok=True)

model_path = os.path.join(out_dir, "xgb_step2_model.joblib")
preds_path = os.path.join(out_dir, "xgb_step2_test_predictions.csv")
meta_path  = os.path.join(out_dir, "xgb_step2_meta.txt")

# Save model and predictions
joblib.dump(final_model, model_path)


preds_df = pd.DataFrame({
    "y_true": y_test.values,
    "y_proba": proba,
    "y_pred": pred
})
preds_df.to_csv(preds_path, index=False)

with open(meta_path, "w") as f:
    f.write(f"Saved: {datetime.datetime.now().isoformat()}\n")
    f.write(f"Data path: {DATA_PATH}\n")
    f.write(f"Train shape: {X_train.shape} | Test shape: {X_test.shape}\n")
    f.write(f"ROC-AUC: {roc_auc:.6f} | PR-AUC: {pr_auc:.6f}\n")
    f.write(f"Scale_pos_weight: {scale_pos_weight}\n")
    f.write(f"Excluded sensitive cols: {EXCLUDE_SENSITIVE}\n")

print("Artifacts saved to:", out_dir)
print(" -", model_path)
print(" -", preds_path)
print(" -", meta_path)
```

## 10. Step 9 Ethnicity Analysis

```python
# =========================
# Step 9 — Fairness w.r.t. Ethnicity (protected attribute)
# =========================
print("\n[Step 9] Fairness evaluation w.r.t. ethnicity (protected attribute)\n")

fair_out_dir = os.path.join(out_dir, "fairness_ethnicity")
os.makedirs(fair_out_dir, exist_ok=True)

# 1) Pick protected attribute column
# Priority: a) an explicit 'ethnicity' categorical column; b) proxy 'Pct_afro_american' (continuous -> bin to groups)
prot_col = None
if "ethnicity" in X_test.columns:
    prot_col = "ethnicity"
elif "Pct_afro_american" in X_test.columns:
    prot_col = "Pct_afro_american"

if prot_col is None:
    raise ValueError(
        "No protected attribute found. Expected 'ethnicity' or 'Pct_afro_american' in the data."
    )

# 2) Build analysis frame (merge predictions with protected attribute on test set)
fair_df = preds_df.copy()
fair_df[prot_col] = X_test[prot_col].values

# 3) If proxy is continuous, discretize into interpretable groups (quantiles)
def make_ethnicity_group(series):
    if pd.api.types.is_numeric_dtype(series):
        # Use 3-quantiles for Low/Med/High (you can switch to 4-quantiles if needed)
        q = pd.qcut(series.fillna(series.median()), q=3, labels=["Low", "Med", "High"])
        return q.astype(str)
    else:
        return series.astype(str)

group_col = "ethnicity_group"
fair_df[group_col] = make_ethnicity_group(fair_df[prot_col])

# 4) Helper metrics
def _rates_for_group(df_g, thr=0.5):
    # predictions at threshold
    y_hat = (df_g["y_proba"].values >= thr).astype(int)
    y = df_g["y_true"].values
    # basic counts
    pos = (y == 1)
    neg = (y == 0)
    # rates
    dp = y_hat.mean() if len(y_hat) else np.nan  # P(ŷ=1)
    # Avoid division by zero
    tpr = (y_hat[pos].mean() if pos.sum() > 0 else np.nan)  # P(ŷ=1|Y=1)
    fpr = (y_hat[neg].mean() if neg.sum() > 0 else np.nan)  # P(ŷ=1|Y=0)
    return dp, tpr, fpr

def compute_group_metrics(df, group_col, thr=0.5):
    rows = []
    for g, dfg in df.groupby(group_col, dropna=False):
        dp, tpr, fpr = _rates_for_group(dfg, thr=thr)
        rows.append({
            group_col: g,
            "count": len(dfg),
            "positive_rate": dfg["y_true"].mean(),
            "DP_rate": dp,
            "EO_TPR": tpr,
            "PredEq_FPR": fpr,
        })
    res = pd.DataFrame(rows).sort_values("count", ascending=False).reset_index(drop=True)
    return res

# 5) Compute metrics at the working threshold (0.5). You can change this if you use a different operating point.
thr0 = 0.5
metrics_table = compute_group_metrics(fair_df, group_col, thr=thr0)

# Choose a reference group:
# - if explicit 'ethnicity': use the largest group as reference
# - if proxy-binned: use 'Low' as reference (least Afro-American share), which matches many use-cases for disparate impact checks
if "ethnicity" in X_test.columns:
    ref_group = metrics_table.iloc[0][group_col]
else:
    ref_group = "Low" if "Low" in metrics_table[group_col].values else metrics_table.iloc[0][group_col]

def add_differences(tbl, ref):
    ref_row = tbl[tbl[group_col] == ref].iloc[0]
    tbl = tbl.copy()
    tbl["DP_diff_vs_ref"] = tbl["DP_rate"] - float(ref_row["DP_rate"])
    tbl["EO_TPR_diff_vs_ref"] = tbl["EO_TPR"] - float(ref_row["EO_TPR"])
    tbl["PredEq_FPR_diff_vs_ref"] = tbl["PredEq_FPR"] - float(ref_row["PredEq_FPR"])
    return tbl

metrics_table = add_differences(metrics_table, ref_group)
print("== Group fairness metrics @ threshold =", thr0, "==")
print(metrics_table)

# Save table
fair_csv_path = os.path.join(fair_out_dir, "fairness_ethnicity_metrics_thr0.50.csv")
metrics_table.to_csv(fair_csv_path, index=False)
print("Saved metrics table ->", fair_csv_path)

# 6) (Optional but recommended) Bootstrap CI for differences vs. reference
USE_BOOTSTRAP = True
N_BOOT = 300       # increase to 1000 for final report if运行时间允许
ALPHA = 0.05
np.random.seed(RANDOM_STATE)

def bootstrap_diff_ci(df, group_col, ref_group, metric="DP_rate", thr=0.5, n_boot=300, alpha=0.05):
    # returns DataFrame with lower/upper CI for (metric_group - metric_ref)
    groups = df[group_col].unique().tolist()
    out = {group_col: [], "stat": [], "diff_mean": [], "ci_low": [], "ci_high": [], "n": []}
    # Precompute index arrays per group
    idx_all = np.arange(len(df))
    by_group_idx = {g: np.where(df[group_col].values == g)[0] for g in groups}
    ref_idx = by_group_idx[ref_group]
    for g in groups:
        gi = by_group_idx[g]
        diffs = []
        for _ in range(n_boot):
            # resample within group and ref group (nonparametric bootstrap)
            gi_bs = np.random.choice(gi, size=len(gi), replace=True)
            ref_bs = np.random.choice(ref_idx, size=len(ref_idx), replace=True)
            dfg = df.iloc[gi_bs]
            dfr = df.iloc[ref_bs]
            # compute rates
            m_g = _rates_for_group(dfg, thr=thr)
            m_r = _rates_for_group(dfr, thr=thr)
            if metric == "DP_rate":
                diff = m_g[0] - m_r[0]
            elif metric == "EO_TPR":
                diff = m_g[1] - m_r[1]
            elif metric == "PredEq_FPR":
                diff = m_g[2] - m_r[2]
            else:
                raise ValueError("Unknown metric for bootstrap CI.")
            diffs.append(diff)
        diffs = np.array(diffs, dtype=float)
        lo, hi = np.nanpercentile(diffs, [100*alpha/2, 100*(1-alpha/2)])
        out[group_col].append(g)
        out["stat"].append(metric)
        out["diff_mean"].append(np.nanmean(diffs))
        out["ci_low"].append(lo)
        out["ci_high"].append(hi)
        out["n"].append(len(gi))
    return pd.DataFrame(out).sort_values("n", ascending=False).reset_index(drop=True)

if USE_BOOTSTRAP:
    boot_tables = []
    for stat in ["DP_rate", "EO_TPR", "PredEq_FPR"]:
        bt = bootstrap_diff_ci(fair_df, group_col, ref_group, metric=stat, thr=thr0, n_boot=N_BOOT, alpha=ALPHA)
        boot_tables.append(bt)
    boot_df = pd.concat(boot_tables, axis=0, ignore_index=True)
    boot_csv_path = os.path.join(fair_out_dir, f"fairness_bootstrap_CI_thr{thr0:.2f}.csv")
    boot_df.to_csv(boot_csv_path, index=False)
    print("Saved bootstrap CIs ->", boot_csv_path)
else:
    boot_df = None

# 7) SP-difference curves with CI bands across thresholds
#    We sweep thresholds in [0.05, 0.95] and plot (DP_group - DP_ref) vs threshold, with bootstrap CI.
def sp_diff_curve(df, group_col, ref_group, thresholds):
    curves = []
    for thr in thresholds:
        met = compute_group_metrics(df, group_col, thr=thr)
        met = add_differences(met, ref_group)
        for _, r in met.iterrows():
            curves.append({"threshold": thr, group_col: r[group_col], "DP_diff_vs_ref": r["DP_diff_vs_ref"], "count": r["count"]})
    return pd.DataFrame(curves)

def sp_diff_curve_boot_ci(df, group_col, ref_group, thresholds, n_boot=200, alpha=0.05):
    # returns long DataFrame: group, threshold, diff_mean, ci_low, ci_high
    groups = df[group_col].unique().tolist()
    rows = []
    ref_idx = np.where(df[group_col].values == ref_group)[0]
    for thr in thresholds:
        # compute point estimate
        met = compute_group_metrics(df, group_col, thr=thr)
        met = add_differences(met, ref_group)
        point = {g: met.loc[met[group_col]==g, "DP_diff_vs_ref"].values[0] for g in groups}
        # bootstrap
        for g in groups:
            gi = np.where(df[group_col].values == g)[0]
            diffs = []
            for _ in range(n_boot):
                gi_bs = np.random.choice(gi, size=len(gi), replace=True)
                ref_bs = np.random.choice(ref_idx, size=len(ref_idx), replace=True)
                dfg = df.iloc[gi_bs]
                dfr = df.iloc[ref_bs]
                m_g = _rates_for_group(dfg, thr=thr)[0]
                m_r = _rates_for_group(dfr, thr=thr)[0]
                diffs.append(m_g - m_r)
            diffs = np.array(diffs, dtype=float)
            lo, hi = np.nanpercentile(diffs, [100*alpha/2, 100*(1-alpha/2)])
            rows.append({
                "threshold": thr,
                group_col: g,
                "DP_diff_mean": point[g],
                "ci_low": lo,
                "ci_high": hi
            })
    return pd.DataFrame(rows)

thr_grid = np.linspace(0.05, 0.95, 19)
sp_curve_df = sp_diff_curve(fair_df, group_col, ref_group, thr_grid)
sp_curve_csv = os.path.join(fair_out_dir, "sp_diff_curve_points.csv")
sp_curve_df.to_csv(sp_curve_csv, index=False)
print("Saved SP-diff curve points ->", sp_curve_csv)

# Bootstrap CI bands (can be time-consuming; n_boot=200 by default here)
SP_BOOT = True
if SP_BOOT:
    sp_ci_df = sp_diff_curve_boot_ci(fair_df, group_col, ref_group, thr_grid, n_boot=200, alpha=ALPHA)
    sp_ci_csv = os.path.join(fair_out_dir, "sp_diff_curve_CI.csv")
    sp_ci_df.to_csv(sp_ci_csv, index=False)
    print("Saved SP-diff curve CIs ->", sp_ci_csv)
else:
    sp_ci_df = None

# 8) Plots
# Bar plot: DP/TPR/FPR by group
plt.figure(figsize=(8, 5))
plt.bar(metrics_table[group_col].astype(str), metrics_table["DP_rate"])
plt.title(f"Demographic Parity by {group_col} @ thr={thr0}")
plt.ylabel("P(ŷ=1)")
plt.xlabel(group_col)
plt.tight_layout()
plt.savefig(os.path.join(fair_out_dir, "bar_DP_rate.png"), dpi=160)
plt.show()

plt.figure(figsize=(8, 5))
plt.bar(metrics_table[group_col].astype(str), metrics_table["EO_TPR"])
plt.title(f"Equal Opportunity (TPR) by {group_col} @ thr={thr0}")
plt.ylabel("TPR = P(ŷ=1|Y=1)")
plt.xlabel(group_col)
plt.tight_layout()
plt.savefig(os.path.join(fair_out_dir, "bar_EO_TPR.png"), dpi=160)
plt.show()

plt.figure(figsize=(8, 5))
plt.bar(metrics_table[group_col].astype(str), metrics_table["PredEq_FPR"])
plt.title(f"Predictive Equality (FPR) by {group_col} @ thr={thr0}")
plt.ylabel("FPR = P(ŷ=1|Y=0)")
plt.xlabel(group_col)
plt.tight_layout()
plt.savefig(os.path.join(fair_out_dir, "bar_PredEq_FPR.png"), dpi=160)
plt.show()

# SP-difference curves with CI bands
if sp_ci_df is not None:
    for g in sorted(sp_ci_df[group_col].unique()):
        if g == ref_group:
            continue
        sub = sp_ci_df[sp_ci_df[group_col] == g].sort_values("threshold")
        plt.figure(figsize=(8, 5))
        plt.plot(sub["threshold"], sub["DP_diff_mean"], label=f"{g} vs {ref_group}")
        plt.fill_between(sub["threshold"], sub["ci_low"], sub["ci_high"], alpha=0.2, label="95% CI")
        plt.axhline(0.0, linestyle="--", linewidth=1)
        plt.xlabel("Decision threshold")
        plt.ylabel("DP difference = P(ŷ=1|g) - P(ŷ=1|ref)")
        plt.title(f"SP-difference curve with 95% CI — {g} vs {ref_group}")
        plt.legend()
        plt.tight_layout()
        fname = f"sp_diff_curve_{g}_vs_{ref_group}.png".replace(" ", "_")
        plt.savefig(os.path.join(fair_out_dir, fname), dpi=160)
        plt.show()

print("\n[Step 9] Fairness evaluation finished.")
print("Artifacts:")
print(" - Group metrics:", fair_csv_path)
if USE_BOOTSTRAP:
    print(" - Bootstrap CI:", boot_csv_path)
print(" - SP-difference points:", sp_curve_csv)
if SP_BOOT:
    print(" - SP-difference CI:", sp_ci_csv)
print(" - Plots saved under:", fair_out_dir)
```
